### Embedding and VectorstorDB

```python
import numpy as np
from sentence_transformers import SentenceTransformer # embedding model
import faiss
import uuid # unique id generator for chromadb
from typing import List, Dict, Any, Tuple
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class EmbeddingManager:
    """Manages document embeddings generation using SentenceTransformer"""

    def __init__(self, model_name: str = 'all-MiniLM-L6-v2'):
        self.model_name = model_name
        self.model = None
        self.load_model()
        
    def load_model(self):
        try:
            print(f"Loading embedding model: {self.model_name}")
            self.model = SentenceTransformer(self.model_name) 
            print("Model loaded successfully.Embedding dimension:", {self.model.get_sentence_embedding_dimension()})
        except Exception as e:
            print(f"Error loading model {self.model_name}: {e}")
            raise    
   
    def generate_embeddings(self, texts: List[str]) -> np.ndarray:

        """Generates embeddings for a list of texts."""
        if not self.model:
            raise ValueError("Model is not loaded.")
        print(f"Generating embeddings for {len(texts)} texts.")
        embeddings = self.model.encode(texts, convert_to_numpy=True)
        return embeddings
    
   ##initialize the embedding manager

embedding_manager = EmbeddingManager()
embedding_manager

texts = []
chunk_metadata = []

for pdf_index, pdf_doc in enumerate(pdf_documents):
    # Skip empty pages
    if pdf_doc.page_content.strip() != "":
        texts.append(pdf_doc.page_content)
        # Store metadata including source PDF index
        meta = pdf_doc.metadata.copy() if hasattr(pdf_doc, 'metadata') else {}
        meta["source_pdf"] = f"PDF_{pdf_index}"
        chunk_metadata.append(meta)

print(f"Total chunks to embed: {len(texts)}")

# Generate embeddings
embeddings = embedding_manager.generate_embeddings(texts)
print("Embeddings shape:", embeddings.shape)

#  verify metadata length
assert len(chunk_metadata) == embeddings.shape[0], "Metadata and embeddings count mismatch!"
```

### VectorStore


```python
class VectorStore:
    """Manages a FAISS vector store for document embeddings."""

    def __init__(self, embedding_dim: int):
        self.embedding_dim = embedding_dim
        self.index = faiss.IndexFlatL2(embedding_dim)
        self.id_to_metadata: Dict[int, Dict[str, Any]] = {}
        self.next_id = 0

    def add_embeddings(self, embeddings: np.ndarray, metadatas: List[Dict[str, Any]] = None):
        """Adds embeddings and their metadata to the vector store."""
        # Ensure embeddings are float32
        embeddings = embeddings.astype(np.float32)
        num_embeddings = embeddings.shape[0]

        # Handle missing or shorter metadata
        if metadatas is None:
            metadatas = [{}] * num_embeddings
        elif len(metadatas) < num_embeddings:
            print(f"Warning: Metadata count ({len(metadatas)}) < embeddings count ({num_embeddings}). Filling missing metadata with empty dicts.")
            metadatas += [{}] * (num_embeddings - len(metadatas))

        # Add embeddings to FAISS index
        self.index.add(embeddings)

        # Map embeddings to metadata
        for i in range(num_embeddings):
            self.id_to_metadata[self.next_id] = metadatas[i]
            self.next_id += 1

        print(f"Added {num_embeddings} embeddings to the vector store.")

    def search(self, query_embedding: np.ndarray, top_k: int = 5) -> List[Tuple[Dict[str, Any], float]]:
        """Searches for the top_k most similar embeddings to the query_embedding."""
        # Ensure float32
        query_embedding = query_embedding.astype(np.float32)
        if query_embedding.ndim == 1:
            query_embedding = np.expand_dims(query_embedding, axis=0)

        D, I = self.index.search(query_embedding, top_k)
        results = []
        for dist, idx in zip(D[0], I[0]):
            metadata = self.id_to_metadata.get(idx, {})
            results.append((metadata, dist))
        return results

# Example usage
vector_store = VectorStore(embedding_dim=embedding_manager.model.get_sentence_embedding_dimension())
vector_store
```

### Retriever Pipeline From VectorStore

```python
class RAGretriever:
    """Retrieval-Augmented Generation (RAG) system using vector store and embedding manager."""

    def __init__(self, vector_store: VectorStore, embedding_manager: EmbeddingManager):
        self.vector_store = vector_store
        self.embedding_manager = embedding_manager

    def retrieve(self, query: str, top_k: int = 5, score_threshold: float = None) -> List[Dict[str, Any]]:
        """
        Retrieves the top_k most relevant documents for the given query.
        
        Args:
            query (str): The input query string.
            top_k (int): Number of top documents to retrieve.
            score_threshold (float, optional): Minimum similarity score threshold.
        """
        try:
            print(f"Retrieving documents for query: {query}")
            print(f"Using top_k={top_k}, score_threshold={score_threshold}")

            # Generate embedding for the query
            query_embedding = self.embedding_manager.generate_embeddings([query])

            # Search the vector store
            results = self.vector_store.search(query_embedding, top_k=top_k)

            retrieved_docs = []

            if results:
                for i, (metadata, score) in enumerate(results):
                    # Convert FAISS distance to similarity score (optional, already done in VectorStore)
                    # Filter by score threshold if provided
                    if score_threshold is None or score >= score_threshold:
                        retrieved_docs.append({
                            "id": metadata.get("id", i),
                            "content": metadata.get("content", ""),  # optional: store actual text in metadata
                            "metadata": metadata,
                            "score": score,
                            "rank": i + 1
                        })

                print(f"Retrieved {len(retrieved_docs)} documents after applying score threshold.")
            else:
                print("No documents retrieved.")

            return retrieved_docs

        except Exception as e:
            print(f"Error during retrieval: {e}")
            return []
rag_retriever = RAGretriever(vector_store, embedding_manager)
rag_retriever
```

### Enhanced RAG Pipeline Feature

```python
from typing import List, Dict, Any

def rag_advanced(
    query: str,
    retriever,              
    llm,                    
    top_k: int = 3,
    min_score: float = 0.2,
    return_context: bool = False,
) -> Dict[str, Any]:
    """
    Enhanced RAG pipeline:
    - retrieves top_k chunks with a similarity threshold
    - calls LLM with combined context + query
    - returns answer, sources, confidence, and optional full context
    """

    # 1) Retrieve documents from vector DB
    results = retriever.retrieve(
        query=query,
        top_k=top_k,
    )

    #Apply similarity threshold manually
    results = [
        r for r in results
        if r.get("similarity_score", 0.0) >= min_score
    ]





    # If nothing found, return empty result
    if not results:
        return {
            "answer": "No relevant context found to answer the question.",
            "sources": [],
            "confidence": 0.0,
            "context": "" if return_context else None,
        }

    # 2) Build context string from all retrieved chunks
    context_parts: List[str] = []
    sources: List[Dict[str, Any]] = []

    for doc in results:
        content = doc.get("content", "")
        metadata = doc.get("metadata", {}) or {}
        score = doc.get("similarity_score", 0.0)

        if content:
            context_parts.append(content)

        # Build a rich source record
        source_info = {
            "source": metadata.get("source") or metadata.get("file_path") or "unknown",
            "page_number": metadata.get("page") or metadata.get("page_number"),
            "similarity_score": float(score),
            "preview": content[:300] if content else ""
        }
        sources.append(source_info)

    context = "\n\n".join(context_parts)

    # Safety check
    if not context.strip():
        return {
            "answer": "No relevant context found to answer the question.",
            "sources": sources,
            "confidence": 0.0,
            "context": "" if return_context else None,
        }

    # 3) Compute an overall confidence (e.g., max similarity)
    confidence = max(s["similarity_score"] for s in sources if "similarity_score" in s)

    # 4) Build prompt with context + query
    prompt = (
        "Use the following context to answer the question concisely.\n\n"
        "Context:\n{context}\n\n"
        "Question:\n{query}\n\n"
        "Answer:"
    )

    # 5) Call LLM
    response = llm.invoke(
        prompt.format(
            context=context,
            query=query
        )
    )

    answer_text = response.content if hasattr(response, "content") else str(response)

    # 6) Return structured result
    return {
        "answer": answer_text,
        "sources": sources,
        "confidence": float(confidence),
        "context": context if return_context else None,
    }

result = rag_advanced(
    query="What is python?",
    retriever=rag_retriever,   
    llm=llm,                  
    top_k=3,
    min_score=0.2,
    return_context=True
)

print("ANSWER:\n", result["answer"])
print("\nSOURCES:")
for s in result["sources"]:
    print(s)
print("\nCONFIDENCE:", result["confidence"])

```
